## Анализ пропусков

```python
def sort_missing_value(data):
    return data.isnull().sum().sort_values(ascending=False)[:15]
```

```python
application_train.shape
```

```python
sort_missing_value(application_train)
```

```python
application_test.shape
```

```python
sort_missing_value(application_test)
```
